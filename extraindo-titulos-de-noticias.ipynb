# Extraíndo os Títulos do Globo.com

* Neste código eu estava buscando extrair os títulos das notícias principais do site jornalístico "globo.com", após ler a documentação e ler alguns artigos a respeito da biblioteca BeautifulSoup, consegui extrair os títulos com sucesso. Minha maior dificuldade foi ao fazer a requisição para o site, pois em diversas tentativas aparecia erro, porém ao pesquisar mais sobre a biblioteca "request" consegui concluir a requisição ao site e consequentemente consluir o scrapping do site  

```python
# Importando as bibliotecas necessárias
import requests
from bs4 import BeautifulSoup

# Fazendo a requisição para a página de notícias
url = 'https://www.globo.com/'
response = requests.get(url)
html = response.content

# Criando o objeto Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Buscando todos os elementos que contêm os títulos das notícias
titulos = soup.find_all('h2', class_='post__title')

# Extraindo e imprimindo os títulos
for titulo in titulos:
    print(titulo.text.strip())
```

# Extraíndo os Subtítulos de outro site de Notícias

* Desta vez, busquei conseguir extrair os subtítulos "< h3 >" de outro site famoso de notícias, nesta etapa meu desafio foi conseguir descobrir por meio do "inspecionar" do site, qual era a estrutura dos títulos deste site, após a minha análise descobri quais títulos possuíam a tag "< h3 >" e a classe "card-font-primary card-font-semibold card-text-neutral-low-500 card-text-xxxs card-leading-xs". Diante destes atributos foi possível concluir o scrapping dos subtítulos do site.

```python
# Importando as bibliotecas necessárias
import requests
from bs4 import BeautifulSoup

# Fazendo a requisição para outra página de notícias
url = 'https://www.r7.com/'
response = requests.get(url)
html = response.content

# Criando o objeto Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Buscando todos os elementos que contêm os títulos das notícias
titulos = soup.find_all('h3', class_='card-font-primary card-font-semibold card-text-neutral-low-500 card-text-xxxs card-leading-xs')

# Extraindo e imprimindo os títulos
for titulo in titulos:
    print(titulo.text.strip())
```
